```python
import sklearn as sk
import pandas as pd
import numpy as np
from pymongo import MongoClient
```

```python
#Connecting to mongodb local db, db called JiraRepos
client = MongoClient('localhost', 27017)
db = client.JiraRepos
collection = db.Mojang

#Querying the file mojang
cursor = collection.find()

```

## The dataset for modelling

```python
#query description and summary for 100 issues
cursor = collection.find().limit(100000)  # Reset cursor with limit
issues = list(cursor)  # Convert cursor to list

# Initialize empty lists for each field
row_ids = []
priority_descs = []
priority_ids = []
issue_types = []
descs = []
summaries = []

# Iterate through all issues and collect data
for issue in issues:
    # Use a default value None for any missing fields
    fields = issue.get("fields", {})
    priority = fields.get("priority", {})
    issuetype = fields.get("issuetype", {})
    
    row_ids.append(issue.get("id", ""))
    priority_descs.append(priority.get("name", "Unknown") if priority else "Unknown")
    priority_ids.append(priority.get("id", "") if priority else "")
    issue_types.append(issuetype.get("name", "Unknown") if issuetype else "Unknown")
    descs.append(fields.get("description", ""))
    summaries.append(fields.get("summary", ""))

# Create a dataframe with all collected data
issue_df = pd.DataFrame({
    'id': row_ids,
    'priority': priority_descs,
    'priority_id': priority_ids,
    'issue_type': issue_types,
    'description': descs,
    'summary': summaries
})


#delete unknown priority values
issue_df = issue_df[issue_df['priority'] != 'Unknown']

#lowercase all columns
issue_df = issue_df.apply(lambda x: x.astype(str).str.lower())

#exporting to output.csv
issue_df.to_csv('output.csv', index=False)



```
